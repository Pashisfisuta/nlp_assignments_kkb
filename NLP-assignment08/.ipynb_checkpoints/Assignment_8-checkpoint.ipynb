#### 1.  What is autoencoder?

自编码是一种神经网络的形式。由编码器和解码器组成，是一种利用反向传播算法使得输出值等于输入值的神经网络，它先将输入压缩成潜在空间表征，然后通过这种表征来重构输出。

**编码器：** 将输入压缩成潜在空间表征

**解码器：** 重构来自潜在空间表征的输入

**应用场景：**
1. 数据去噪
2. 进行可视化降维


- 参考网址：<https://zhuanlan.zhihu.com/p/34238979>

#### 2. What are the differences between greedy search and beam search?

**greedy search:** 解码的每一步都选择最可能的单词，最后得到句子的每一个单词都是每一步认为的最合适的单词。缺点是无法保证整个句子的概率最大，即不能保证整个句子最合适。

**beam search:** 每步选取最可能的𝑘个结果，再从最后的𝑘个结果中选取最合适的句子

#### 3. What is the intuition of attention mechanism?

- 用于提升基于RNN（LSTM或GRU）的Encoder + Decoder模型的效果的的机制（Mechanism）

- 一种资源分配模型，在某个特定时刻，你的注意力总是集中在画面中的某个焦点部分，而对其它部分视而不见。

**优点：**
- Attention给模型赋予了区分辨别的能力

- 在机器翻译、语音识别应用中，为句子中的每个词赋予不同的权重，使神经网络模型的学习变得更加灵活（soft），同时Attention本身可以做为一种对齐关系，解释翻译输入/输出句子之间的对齐关系，解释模型到底学到了什么知识，为我们打开深度学习的黑箱，提供了一个窗口。


- 与人类对外界食物的观察机制类似，当人类观察外界事物的时候，一般不会把事物当成一个整体去看，往往倾向于根据需要选择性的去获取被观察事物的某些重要部分。

#### 6. Compared to RNN,  what is the advantage of Transformer ?

- 本质上是基于attention机制，采用了多头注意力机制，对于长句子中相隔较远的也能学习到器相互之间的依赖。

- 不同于RNN，不需要在输入端，添加位置符号，而是使用位置函数来进行位置编码。

- RNN核心是每个输入对应隐层节点，而隐层节点之间形成了线性序列，信息由前向后在隐层之间逐步向后传递。这种线性结构在反向传播，由于路径太长，容易导致梯度消失等问题。

- RNN并行计算能力差

#### 7. Why we use layer normalizaiton instead of batch normalization in Transformer ?

- batch normalization: 是对一批数据，整个进行归一化，可能无法有效提取出有效信息

- layer normalization: 是对一批数据中的一个样本沿着词向量的方向标准化，可以提取每个词向量中的重要信息。

#### 8. Why we need position embedding in Transformer ?

- 因为transformer不同于RNN，不需要拆分序列，因此输入序列之间无法获取前后位置信息，所以引入位置词向量。

#### 9. Briefly describe what is self-attention and what is multi-head attention?

- self-attention: 计算每个单词和其他所有单词之间的联系，通过计算Q，K，V，表示上下文的信息。

- multi-head attention: 多个Self-Attention的结合，每个head学习到在不同表示空间中的特征。

#### 12. What is masked language model in BERT ?

- 是指在训练的时候随即从输入预料上mask掉一些单词，然后通过的上下文预测该单词

- 在BERT模型中，15%的words会被随机Mask掉。在训练模型时，一个句子会被多次喂到模型中用于参数学习，但是Google并没有在每次都mask掉这些单词，而是在确定要Mask掉的单词之后，80%的时候会直接替换为[Mask]，10%的时候将其替换为其它任意单词，10%的时候会保留原始Token。

- 这样做的目的可以实现预训练的深度双向表示。
